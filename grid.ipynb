```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from tqdm.notebook import tqdm_notebook
```

```python
nx = 4
ny = 2
nz = 1

file1 = open('x.dat', 'r')
lines = file1.readlines()

x_cord = []
for line in tqdm_notebook(lines):
    items= line.split(" ")
    for item in items:
        if item.find('*') >= 0:
            x= item.split('*')
            for j in range(int(x[0])):
                x_cord.append(float(x[1]))
        elif item:
            x_cord.append(float(item))
            
file1.close()
```

```python
file1 = open('y.dat', 'r')
lines = file1.readlines()

y_cord = []
for line in tqdm_notebook(lines):
    items= line.split(" ")
    for item in items:
        if item.find('*') >= 0:
            x= item.split('*')
            for j in range(int(x[0])):
                y_cord.append(float(x[1]))
        elif item:
            y_cord.append(float(item))
            
file1.close()
```

```python
file1 = open('z.dat', 'r')
lines = file1.readlines()

z_cord = []
for line in tqdm_notebook(lines):
    items= line.split(" ")
    for item in items:
        if item.find('*') >= 0:
            x= item.split('*')
            for j in range(int(x[0])):
                z_cord.append(float(x[1]))
        elif item:
            z_cord.append(float(item))

z_cord = list(map(lambda z: z * -1, z_cord))            
file1.close()
```

```python
file1 = open('por.dat', 'r')
lines = file1.readlines()

por = []
for line in tqdm_notebook(lines):
    items= line.split(" ")
    for item in items:
        if item.find('*') >= 0:
            x= item.split('*')
            for j in range(int(x[0])):
                por.append(float(x[1]))
        elif item:
            por.append(float(item))
            
file1.close()
```

```python
index = 0
arr_ijk = np.ones([nx*ny*nz, 3], dtype = int)*-99
for k in tqdm_notebook(range(1,nz+1)):
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            arr_ijk[index,:] = [i,j,k]
            index = index + 1

df_ijk = pd.DataFrame(data = arr_ijk,  columns = ['I','J','K'])
```

```python
ai = 0
values = x_cord
NWT_x=[]
NET_x=[]
SWT_x=[]
SET_x=[]
NWB_x=[]
NEB_x=[]
SWB_x=[]
SEB_x=[]

for k in tqdm_notebook(range(1,nz+1)):
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWT_x.append(values[ai])
            ai=ai+1
            NET_x.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWT_x.append(values[ai])
            ai=ai+1
            SET_x.append(values[ai])
            ai=ai+1
            
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWB_x.append(values[ai])
            ai=ai+1
            NEB_x.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWB_x.append(values[ai])
            ai=ai+1
            SEB_x.append(values[ai])
            ai=ai+1

arr_x= np.hstack((np.array(NWT_x).reshape(-1,1), np.array(NET_x).reshape(-1,1), np.array(SWT_x).reshape(-1,1), 
                  np.array(SET_x).reshape(-1,1), np.array(NWB_x).reshape(-1,1), np.array(NEB_x).reshape(-1,1),
                  np.array(SWB_x).reshape(-1,1), np.array(SEB_x).reshape(-1,1)))
           
df_x = pd.DataFrame(data = arr_x,  columns = ['NWT_x','NET_x','SWT_x','SET_x','NWB_x','NEB_x','SWB_x','SEB_x'])
```

```python
ai = 0
values = y_cord
NWT_y=[]
NET_y=[]
SWT_y=[]
SET_y=[]
NWB_y=[]
NEB_y=[]
SWB_y=[]
SEB_y=[]

for k in tqdm_notebook(range(1,nz+1)):
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWT_y.append(values[ai])
            ai=ai+1
            NET_y.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWT_y.append(values[ai])
            ai=ai+1
            SET_y.append(values[ai])
            ai=ai+1
            
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWB_y.append(values[ai])
            ai=ai+1
            NEB_y.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWB_y.append(values[ai])
            ai=ai+1
            SEB_y.append(values[ai])
            ai=ai+1

arr_y= np.hstack((np.array(NWT_y).reshape(-1,1), np.array(NET_y).reshape(-1,1), np.array(SWT_y).reshape(-1,1), 
                  np.array(SET_y).reshape(-1,1), np.array(NWB_y).reshape(-1,1), np.array(NEB_y).reshape(-1,1),
                  np.array(SWB_y).reshape(-1,1), np.array(SEB_y).reshape(-1,1)))
           
df_y = pd.DataFrame(data = arr_y,  columns = ['NWT_y','NET_y','SWT_y','SET_y','NWB_y','NEB_y','SWB_y','SEB_y'])
```

```python
ai = 0
values = z_cord
NWT_z=[]
NET_z=[]
SWT_z=[]
SET_z=[]
NWB_z=[]
NEB_z=[]
SWB_z=[]
SEB_z=[]

for k in tqdm_notebook(range(1,nz+1)):
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWT_z.append(values[ai])
            ai=ai+1
            NET_z.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWT_z.append(values[ai])
            ai=ai+1
            SET_z.append(values[ai])
            ai=ai+1
            
    for j in range(1,ny+1):
        for i in range(1,nx+1):
            NWB_z.append(values[ai])
            ai=ai+1
            NEB_z.append(values[ai])
            ai=ai+1
        for i in range(1,nx+1):
            SWB_z.append(values[ai])
            ai=ai+1
            SEB_z.append(values[ai])
            ai=ai+1

arr_z= np.hstack((np.array(NWT_z).reshape(-1,1), np.array(NET_z).reshape(-1,1), np.array(SWT_z).reshape(-1,1), 
                  np.array(SET_z).reshape(-1,1), np.array(NWB_z).reshape(-1,1), np.array(NEB_z).reshape(-1,1),
                  np.array(SWB_z).reshape(-1,1), np.array(SEB_z).reshape(-1,1)))
           
df_z = pd.DataFrame(data = arr_z,  columns = ['NWT_z','NET_z','SWT_z','SET_z','NWB_z','NEB_z','SWB_z','SEB_z'])
```

```python
df_por = pd.DataFrame(data = np.array(por).reshape(-1,1),  columns = ['Porosity'])
df= pd.concat([df_ijk,df_x,df_y,df_z,df_por], axis=1)

data = df.to_numpy()
```

```python
rw=0
x1=[data[rw,6],data[rw,5],data[rw,3],data[rw,4],data[rw,10],data[rw,9],data[rw,7],data[rw,8]]
y1=[data[rw,14],data[rw,13],data[rw,11],data[rw,12],data[rw,18],data[rw,17],data[rw,15],data[rw,16]]
z1=[data[rw,22],data[rw,21],data[rw,19],data[rw,20],data[rw,26],data[rw,25],data[rw,23],data[rw,24]]
por1=(np.ones(8)*data[rw,27]).tolist()
i1 = [7, 0, 0, 0, 4, 4, 6, 6, 4, 0, 3, 2]
j1 = [3, 4, 1, 2, 5, 6, 5, 2, 0, 1, 6, 3]
k1 = [0, 7, 2, 3, 6, 7, 1, 1, 5, 5, 7, 6]

for rw in tqdm_notebook(range(1,df.shape[0])):
    x1.extend([data[rw,6],data[rw,5],data[rw,3],data[rw,4],data[rw,10],data[rw,9],data[rw,7],data[rw,8]])
    y1.extend([data[rw,14],data[rw,13],data[rw,11],data[rw,12],data[rw,18],data[rw,17],data[rw,15],data[rw,16]])
    z1.extend([data[rw,22],data[rw,21],data[rw,19],data[rw,20],data[rw,26],data[rw,25],data[rw,23],data[rw,24]])
    por1.extend((np.ones(8)*data[rw,27]).tolist())
    i1.extend((np.array([7, 0, 0, 0, 4, 4, 6, 6, 4, 0, 3, 2])+8*rw).tolist())
    j1.extend((np.array([3, 4, 1, 2, 5, 6, 5, 2, 0, 1, 6, 3])+8*rw).tolist())
    k1.extend((np.array([0, 7, 2, 3, 6, 7, 1, 1, 5, 5, 7, 6])+8*rw).tolist())
```

```python
por2=[str(x) for x in por1]

fig = go.Figure(data=[
go.Mesh3d(
# 8 vertices of a cube
x=x1,
y=y1,
z=z1,
# set colour scale
colorscale=[[0, 'white'],[0.5, 'blue'],[1, 'green']],
cmax = 1,
cmin = 0,
# Intensity of each vertex, which will be interpolated and color-coded
intensity = por1,
# i, j and k give the vertices of triangles
i = i1,
j = j1,
k = k1,
showscale=True,
hoverinfo='text+x+y+z',
hovertext = por2
    )
])
             
fig.update_layout(width=1000,height=600)

fig.show()
```
